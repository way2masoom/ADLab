```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Given data points
points = np.array([[0.40, 0.53], [0.22, 0.38], [0.35, 0.32], [0.26, 0.19], [0.08, 0.41], [0.45, 0.30]])

# Convert to DataFrame
data = pd.DataFrame(points, columns=['X', 'Y'])
```

```python
# Applying K-Means with 2 clusters (assuming K=2)
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
data['Cluster'] = kmeans.fit_predict(points)
```

```python
# Plot the clusters in 2D space
plt.figure(figsize=(8,6))
sns.scatterplot(x=data['X'], y=data['Y'], hue=data['Cluster'], palette='viridis', s=100, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centroids')
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("K-Means Clustering of Given Points")
plt.legend()
plt.show()
```

```python
# Display cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```
