```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Given data points
points = np.array([[2, 10], [2, 5], [8, 4], [5, 8], [7, 5], [6, 4], [1, 2], [4, 9]])

# Convert to DataFrame
data = pd.DataFrame(points, columns=['X', 'Y'])
```

```python
# Applying K-Means with 3 clusters (assuming K=3)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
data['Cluster'] = kmeans.fit_predict(points)
```

```python
# Plot the clusters in 2D space
plt.figure(figsize=(8,6))
sns.scatterplot(x=data['X'], y=data['Y'], hue=data['Cluster'], palette='viridis', s=100, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centroids')
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("K-Means Clustering of Given Points")
plt.legend()
plt.show()
```

```python
# Display cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```
